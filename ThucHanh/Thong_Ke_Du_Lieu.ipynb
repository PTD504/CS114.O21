```python
import pandas as pd
import os, glob, csv

dir = 'drive/MyDrive/CS114/Public/'
# dir = 'dir/'

yamaha_path = dir + 'Yamaha'
vinfast_path = dir + 'VinFast'
suzuki_path = dir + 'Suzuki'
others_path = dir + 'Others'
honda_path = dir + 'Honda'

'''
mssv1-mssv2.Honda.1.jpg
MSSVs[mssv] = {'Yamaha': So_luong, 'VinFast': So_luong, 'Suzuki': So_luong, 'Others': So_luong, 'Honda': So_luong}
'''

automakers = {yamaha_path: 'Yamaha', vinfast_path: 'VinFast', suzuki_path: 'Suzuki', others_path: 'Others', honda_path: 'Honda'}
MSSVs = {}

for path, name in automakers.items():
  for _, _, files in os.walk(path):
    for file_name in files:
      if file_name.endswith(('.jpg', '.jpeg', '.heic', '.png')):
        li = file_name.split('.')
        # Chưa đúng định dạng
        if len(li) != 4:
          continue
        # Ảnh không đúng với tên thư mục (tên hãng xe mà thư mục quy định)
        if li[-3] != name:
          continue
        mssvs = li[0].split('-')

        for mssv in mssvs:
          if mssv in MSSVs:
            if name in MSSVs[mssv]:
              MSSVs[mssv][name] += 1
            else:
              MSSVs[mssv][name] = 1
          else:
            MSSVs[mssv] = {}
            MSSVs[mssv][name] = 1

'''
file1: mssv, All, So_luong
file2: mssv, Hang_xe, So_luong
'''

# Tổng hợp lại dữ liệu để ghi vào file csv
data_file1 = []
data_file2 = []

for key, items in MSSVs.items():
  num_all = 0.0

  for keyD, itemsD in items.items():
    num_all += itemsD
    data_file2.append([key, keyD, itemsD])

  # Những hãng xe không có ảnh nào của sinh viên có MSSV là key
  if len(items) < 5:
    for _, item in automakers.items():
      if item not in items:
        data_file2.append([key, item, 0.0])
  data_file1.append([key, 'All', num_all])

file1 = 'MotocycleDataset-1.csv'
file2 = 'MotocycleDataset-2.csv'
# file1 = dir + 'MotocycleDataset-1.csv'
# file2 = dir + 'MotocycleDataset-2.csv'

# Write file
with open(file1, 'w', newline='') as file:
  writer = csv.writer(file)
  writer.writerows(data_file1)

with open(file2, 'w', newline='') as file:
  writer = csv.writer(file)
  writer.writerows(data_file2)

###
df1 = pd.read_csv(file1, header=None)
df2 = pd.read_csv(file2, header=None)

col1_name = ['MSSV', 'All', 'Số lượng']
col2_name = ['MSSV', 'Hãng xe', 'Số lượng']

df1.columns = col1_name
df2.columns = col2_name

print(f'MotocycleDataset-1.csv: \n\n{df1}\n\n\n')
print(f'MotocycleDataset-2.csv: \n\n{df2}')
```
